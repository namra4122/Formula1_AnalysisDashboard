<h2> Import Libs

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
```

```python
race_api = f"https://ergast.com/api/f1/2023.json"
race_data = []

races = requests.get(race_api).json()["MRData"]["RaceTable"]["Races"]

for race in races:
    race_data.append({
        "raceName" : race["raceName"],
        "circuitid" : race["Circuit"]["circuitId"]
    })

race_df = pd.DataFrame(race_data)  
```

<h2>Driver Data Extraction

```python
api_url = 'http://ergast.com/api/f1/drivers.json?limit=1000'
response = requests.get(api_url)
data = response.json()
driver_data = []
drivers = data['MRData']['DriverTable']['Drivers']
for driver in drivers:
    driver_data.append({
        'driver_id': driver['driverId'],
        'driver_name': driver['givenName'] + ' ' + driver['familyName'],
        'date_of_birth': driver['dateOfBirth'],
        'nationality': driver['nationality']
    })
```

```python
driver_df = pd.DataFrame(driver_data)

driver_df.to_dict('records')
```

<h2> Constructor Data Extraction

```python
api_url = 'https://ergast.com/api/f1/constructors.json?limit=500'
response = requests.get(api_url)
data = response.json()
constructor_data = []
constructors = data['MRData']['ConstructorTable']['Constructors']
for constructor in constructors:
    constructor_data.append({
        'constructor_id': constructor['constructorId'],
        'constructor_name': constructor['name'],
        'nationality': constructor['nationality']
    })
```

```python
constructor_df = pd.DataFrame(constructor_data)
constructor_df
```

<h2> Circuit Data Extraction

```python
api_url = 'https://ergast.com/api/f1/circuits.json?limit=100'
response = requests.get(api_url)
data = response.json()
circuit_data = []
circuits = data['MRData']['CircuitTable']['Circuits']
for circuit in circuits:
    circuit_data.append({
        'circuit_id': circuit['circuitId'],
        'circuit_name': circuit['circuitName'],
        'circuit_country': circuit['Location']['country']
    })
```

```python
circuit_df = pd.DataFrame(circuit_data)
circuit_df
```

```python
circuit_df.to_csv("circuit.csv")
```

<h2> Season Data Extraction

```python
api_url = 'https://ergast.com/api/f1/seasons.json?limit=100'
response = requests.get(api_url)

data = response.json()
season_data = []

seasons = data['MRData']['SeasonTable']['Seasons']

for season in seasons:
    # print(season)
    season_data.append(season['season'])
```

<h2> Race Data Extraction

```python
race_data = []

for i,row in season_df.iterrows():
    api_url = 'https://ergast.com/api/f1/%d/results.json?limit=1500'%(row['season_year'])
    
    response = requests.get(api_url)
    data = response.json()
    races = data['MRData']['RaceTable']['Races']
    for race in races:
        race_data.append({
            'race_season' : int(race['season']),
            'race_round' : int(race['round']),
            'race_name' : race['raceName'],
            'race_circuitName' : race['Circuit']['circuitName'],
            'race_circuitLocation' : race['Circuit']['Location']['country'],
            'race_date' : race['date'],
            'podium_1' : race['Results'][0]['Driver']['givenName'] + ' ' + race['Results'][0]['Driver']['familyName'],
            'podium_2' : race['Results'][1]['Driver']['givenName'] + ' ' + race['Results'][1]['Driver']['familyName'],
            'podium_3' : race['Results'][2]['Driver']['givenName'] + ' ' + race['Results'][2]['Driver']['familyName']
            
        })
```

```python
race_df = pd.DataFrame(race_data)
```
